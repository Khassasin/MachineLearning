```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x1 = np.linspace(-1, 1, 500)
x2 = np.linspace(-1, 1, 500)
x1, x2 = np.meshgrid(x1, x2)
y = x1 ** 2 / (1 + np.exp((-2) * x2))
df = pd.DataFrame({'x1': x1.ravel(), 'x2': x2.ravel(), 'y': y.ravel()})
df.to_csv('test.csv')

```

```python
new_df = pd.read_csv('test.csv')
print('x1: min', x1.min(), 'max', x1.max())
print('x2: min', x2.min(), 'max', x2.max())
print('y: min', y.min(), 'max', y.max())

const = 50
plt.plot(x1[const, :], y[const, :])
plt.show()
plt.plot(x2[:, const], y[:, const])
plt.show()
```

```python
condition = (new_df['x1'] < new_df['x1'].mean()) | (new_df['x2'] < new_df['x2'].mean())
filtered_df = df[condition]
filtered_df.to_csv('less.csv')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x1, x2, y, cmap='inferno')
plt.show()
```
